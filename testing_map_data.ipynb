```python
postcodes = postcode_lookup[['postcode', 'latitude', 'longitude']]
```

```python
postcodes['postcode']
```

```python
postcodes[postcodes['postcode'] == 'CM11 2RE']
```

```python
london_epd_coords = london_epd.merge(postcodes, left_on='POSTCODE', right_on='postcode', how='left')
```

```python
london_epd_coords
```

```python
from shapely.geometry import Point
from geopandas import GeoDataFrame

geometry = [Point(xy) for xy in zip(london_epd_coords.longitude, london_epd_coords.latitude)]
epd2 = london_epd_coords.drop(['longitude', 'latitude'], axis=1)
epd_gdf = GeoDataFrame(epd2, crs="EPSG:4326", geometry=geometry)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
shapefile.to_crs(epsg=4326).plot(ax=ax, color='gray')
epd_gdf.plot(ax=ax)
ax.set_title('Prescriptions')
```

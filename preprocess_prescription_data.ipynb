```python
def zip_to_csv(zip_path, csv_path, save_path):
    archive = zipfile.ZipFile(zip_path, 'r')
    zipped_csv = pd.read_csv(archive.open(csv_path))
    london_csv = zipped_csv[zipped_csv['REGIONAL_OFFICE_NAME'] == 'LONDON']
    london_antidepressants_csv = london_csv[london_csv['BNF_CODE'].str.startswith('0403', na=False)]
    london_antidepressants_csv.to_csv(save_path)
```
